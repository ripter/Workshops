# LLM Talk

Getting started with LLMs using the Hugging Face `transformers` library.

```python
import transformers

print(f"transformers version: {transformers.__version__}")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
#tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
#tokenizer = AutoTokenizer.from_pretrained("t5-small")


#text = "A Tokenizer is a model that converts text into number values called 'tokens'."
text = "A quick brown fox jumps over the LLM performing complex calulations to save the world."
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text)

#print("Tokens:", tokens)
#print("Token IDs:", token_ids)
for t, i in zip(tokens, token_ids):
    print(f"{t} -> {i}")
```

```python
decoded = tokenizer.decode(token_ids)
print(decoded)
```
